```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
def con_query(query):
    with engine.connect() as con:
        x = query
        rs = con.execute(x)
        df = pd.DataFrame(rs.fetchall())
        df.columns = rs.keys()
    return df
```

```python
query1 = 'SELECT * \
FROM Facilities' 
```

```python
con_query(query1)
```

```python
query2 = 'SELECT * \
FROM Members'
```

```python
con_query(query2)
```

```python
query3 = 'SELECT * \
FROM Bookings \
ORDER BY starttime'
```

```python
con_query(query3)
```

```python
query4 = 'SELECT facility, SUM(cost) AS revenue \
FROM (SELECT f.name AS facility, \
CASE WHEN b.memid = 0 THEN b.slots * f.guestcost \
ELSE b.slots * f.membercost END AS cost \
FROM Bookings AS b \
LEFT JOIN Facilities AS f ON b.facid = f.facid) AS costs \
GROUP BY facility \
HAVING revenue < 1000 \
ORDER BY revenue'
```

```python
con_query(query4)
```

```python
query5 = 'SELECT m1.surname || ", " || m1.firstname AS lastfirst, m2.surname || ", " || m2.firstname AS recommendedby \
FROM Members AS m1 \
LEFT JOIN Members AS m2 ON m1.recommendedby = m2.memid \
WHERE m1.memid != 0 \
ORDER BY m1.surname, m1.firstname'
```

```python
con_query(query5)
```

```python
query6 = 'SELECT f.name as facility, m.firstname || " " || m.surname AS fullname, SUM(b.slots) AS slotsbooked \
FROM Bookings AS b \
LEFT JOIN MEMBERS as m ON b.memid = m.memid \
LEFT JOIN FACILITIES as f ON b.facid = f.facid \
WHERE b.memid != 0 \
GROUP BY facility, b.memid'
```

```python
con_query(query6)
```

```python
query7 = 'SELECT f.name AS facility, strftime("%m", starttime) as month, SUM(b.slots) AS slotsbooked \
FROM Bookings AS b \
INNER JOIN Facilities AS f ON b.facid = f.facid \
WHERE b.memid != 0 \
GROUP BY facility, month'
```

```python
con_query(query7)
```
